```python
import pandas as pd

df = pd.read_csv('../data/train.csv')
```

```python
binary_features = [
    'has_code_block', 'has_error_keywords',
    'has_attempt_signal', 'has_expected_vs_actual',
    'has_question_mark'
]
```

```python
from src.features import extract_features

sample = df.sample(5000, random_state=42)
feature_data = sample.apply(
    lambda row: extract_features(
        row['Title'], row['Body'], row['Tags']
    ), axis=1
)
feature_df = pd.DataFrame(list(feature_data))
feature_df['Y'] = sample['Y'].values

# Show percentage of 1s per class per feature
for feat in binary_features:
    print(f"\n{feat}:")
    print(feature_df.groupby('Y')[feat].mean().round(3))
```

```python
import pandas as pd

df = pd.read_csv('../data/train.csv')

# Loop through each label group and print samples
for label in ['HQ', 'LQ_CLOSE', 'LQ_EDIT']:
    
    print(f"\n{'='*60}")
    print(f"LABEL: {label}  ({len(df[df['Y']==label])} total questions)")
    print(f"{'='*60}")
    
    # Get 3 samples from this label
    samples = df[df['Y'] == label].sample(3, random_state=42)
    
    for i, (_, row) in enumerate(samples.iterrows()):
        print(f"\n--- Sample {i+1} ---")
        print(f"TITLE: {row['Title']}")
        print(f"BODY (first 400 chars):\n{str(row['Body'])[:400]}")
        print()
```
